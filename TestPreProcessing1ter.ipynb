```python
import geopy.distance
def calc_orthodromic(row):
    try:
        return geopy.distance.geodesic(row["XY"], row["XY_cas"]).km
    except:
        return np.NaN

Final_1['XY'] = list(zip(Final_1["lat"], Final_1["long"]))
Final_1['XY_cas'] = list(zip(Final_1["Latitude_y"], Final_1["Longitude_y"]))

Final_1['distance'] = Final_1.apply(calc_orthodromic, axis=1)
```

```python
Final_1.drop(['NumStationsWithPumpsAttending','NumPumpsAttending','PumpCount','Latitude_x',
             'Longitude_x','Latitude_y','Longitude_y','long','lat',
             'XY','XY_cas'], axis=1,inplace=True)
Final_1.head()
```

```python
Final_1['index_col'] = Final_1.index  
def mean_without_self(group):     
    return group.apply(lambda row: group[group['index_col'] != row['index_col']]['TravelTimeSeconds'].mean(), axis=1)  
new_column = Final_1.groupby(['DeployedFromStation_Name', 'HourOfCall']).apply(mean_without_self).reset_index(level=0, drop=True)   
Final_1['average_time_of_arrival_Station_Hour'] = new_column.reset_index(level=0, drop=True)
Final_1.drop(['index_col'], axis=1,inplace=True)
Final_1.head()
```
