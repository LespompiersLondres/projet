```python
Final_1b['DateOfCall_bis'] = pd.to_datetime(Final_1b['DateOfCall_bis'])
Final_1b=Final_1b.dropna()
Final_1b.info()
```

```python
from sklearn import linear_model,preprocessing
from sklearn.model_selection import train_test_split
```

```python
Final_1b['TravelTimeClass']=pd.qcut(Final_1b['TravelTimeSeconds'],4,labels=[0,1,2,3])
```

```python
data=Final_1b.drop(['TravelTimeSeconds','TurnoutTimeSeconds','AttendanceTimeSeconds','Resource_Code','IncidentNumber',
                    'DateOfCall_bis','TravelTimeClass'], axis=1)
target=Final_1b['TravelTimeClass']
```

```python
X_train,X_test,y_train,y_test=train_test_split(data,target,test_size=0.2,random_state=66)
```

```python
X_train_scaled=preprocessing.scale(X_train)
scaler=preprocessing.StandardScaler().fit(X_train)
X_train_scaled=scaler.transform(X_train)
scaler=preprocessing.StandardScaler().fit(X_test)
X_test_scaled=scaler.transform(X_test)
```

```python
clf=linear_model.LogisticRegression()
clf.fit(X_train_scaled,y_train)
y_pred=clf.predict(X_test_scaled)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
print(cm)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
clf.score(X_test,y_test)
```

```python
data=Final_1b.drop(['TravelTimeSeconds','TurnoutTimeSeconds','AttendanceTimeSeconds','Resource_Code','IncidentNumber',
                    'DateOfCall_bis','TravelTimeClass'], axis=1)
target=Final_1b['TravelTimeSeconds']
```

```python
X_train,X_test,y_train,y_test=train_test_split(data,target,test_size=0.2,random_state=66)
```

```python
X_train_scaled=preprocessing.scale(X_train)
scaler=preprocessing.StandardScaler().fit(X_train)
X_train_scaled=scaler.transform(X_train)
scaler=preprocessing.StandardScaler().fit(X_test)
X_test_scaled=scaler.transform(X_test)
```

```python
from sklearn import neighbors,model_selection,ensemble,svm
```

```python
knn=neighbors.KNeighborsClassifier()
parametres={'n_neighbors':range(2,40)}
grid_knn=model_selection.GridSearchCV(estimator=knn,param_grid=parametres)
grid_knn.fit(X_train_scaled,y_train)
```

```python
print('Meilleurs paramÃ¨tres:',grid_knn.best_params_)
print(grid_knn.score(X_test_scaled,y_test))
y_pred=grid_knn.predict(X_test_scaled)
pd.crosstab(y_test,y_pred)
```

```python
svm_clf=svm.SVC(probability=True)
parametres={'C':[0.1,1,10,30],'kernel':['rbf','linear','sigmoid']}
grid_svm=model_selection.GridSearchCV(estimator=svm_clf,param_grid=parametres)
grid_svm.fit(X_train_scaled,y_train)
grid_svm.best_params_
print(grid_svm.score(X_test_scaled,y_test))
y_pred_svm=grid_svm.predict(X_test_scaled)
pd.crosstab(y_test,y_pred_svm)
```

```python
rf=ensemble.RandomForestClassifier()
grid_rf=model_selection.GridSearchCV(estimator=rf, param_grid={'max_features':["squrt","log2",None]})
grid_rf.fit(X_train_scaled,y_train)
grid_rf.best_params_
print(grid_rf.score(X_test_scaled,y_test))
y_pred_rf=grid_rf.predict(X_test_scaled)
pd.crosstab(y_test,y_pred_rf)
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import f1_score
```
